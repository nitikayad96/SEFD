```python
import numpy as np
import struct
import sys
import matplotlib.pyplot as plt
from scipy import optimize
from astropy import modeling
```

```python
# modified from Vikram's code, samp.py
# change out.dat path to run on different machine
data = open('/home/nitikayad96/GradSchool/Research/DSA/out.dat','rb').read()
n = int(len(data)/4)
d = np.asarray(struct.unpack(str(n)+"f",data))
nt = int(len(d)/int((2048/8)))

spectra = d.reshape((nt,int(2048/8)))
```

```python
print(spectra.shape)
```

```python
# create arrays for times and freq channels

tsamp = 8.4e-3 # s
times = np.array(np.arange(spectra.shape[0]))*tsamp

f0 = 1530 # MHz
fe = 1280 # MHz
nchanns = spectra.shape[1]
fsamp = (f0-fe)/nchanns

fchanns = np.array(np.arange(nchanns))
fchanns = f0 - fchanns*fsamp
```

```python
# only use subset of data in the frequency channels described
# in the memo

intf = np.argwhere((fchanns>1403) & (fchanns<1461.6))
intf = np.ravel(intf)
```

```python
# integrate power across this slice of frequency channels
spectra_ = spectra[:,intf]
spec_int = np.mean(spectra_,1)
```

```python
# re-bin samples by average sets of 32 samples

mean_times = np.mean(times.reshape(-1, 32), axis=1)
mean_spec = np.mean(spec_int.reshape(-1, 32), axis=1)
```

```python
# By inspection, only select times where it looks like there are CygA transits

cyga = np.argwhere((mean_times>1100) & (mean_times<1550))
db = 10*np.log10(mean_spec)

times_cyga = np.ravel(mean_times[cyga])
spec_cyga =  np.ravel(mean_spec[cyga])
db_cyga = np.ravel(db[cyga])

# plot power in raw units + relative dB
plt.figure(0)
plt.plot(times_cyga, spec_cyga, 'm')
plt.xlabel('Time (s)');
plt.ylabel('Power');
plt.title('Cyg A Transits');

plt.figure(1)
plt.plot(times_cyga, db_cyga)
plt.xlabel('Time (s)');
plt.ylabel('Power (relative dB)');
plt.title('Cyg A Transits');

# find background regions by taking first derivative and only counting regions where
# the deriative is within 1 std dev of mean
# then exclude peaks (which will also have f'=0)

spec_diff1 = np.gradient(spec_cyga)
bkg = np.where(np.abs(spec_diff1) <= np.mean(spec_diff1) + np.std(spec_diff1))
bkg_times = times_cyga[bkg]
bkg_spec = spec_cyga[bkg]

not_peaks = np.where(bkg_spec < np.mean(bkg_spec) + 2*np.std(bkg_spec))
bkg_times = bkg_times[not_peaks]
bkg_spec = bkg_spec[not_peaks]

plt.figure(3)
plt.plot(times_cyga, spec_cyga, 'm')
plt.plot(bkg_times, bkg_spec,'g.', markersize=5)
plt.xlabel('Time (s)');
plt.ylabel('Power');
plt.title('Cyg A Transits');
```

```python
# interpolate between gaps to get background prediction

bkg_spec_total = np.interp(times_cyga,bkg_times, bkg_spec)

plt.figure(4)
plt.plot(times_cyga, spec_cyga, 'm')
plt.plot(times_cyga, bkg_spec_total,'g.', markersize=5)
plt.xlabel('Time (s)');
plt.ylabel('Power');
plt.title('Cyg A Transits');
```

```python
# remove bkg from light curve by bringing areas with no source
# down to cold sky power
cold = np.min(bkg_spec_total)
spec_fin = spec_cyga - (bkg_spec_total - cold)

plt.figure(5)
#plt.plot(times_cyga, spec_cyga, 'm')
plt.plot(times_cyga, spec_fin, 'b')
plt.xlabel('Time (s)');
plt.ylabel('Power');
plt.title('Background Corrected Cyg A Transits');
```

```python
# get SEFD using eq. given in memo

Scyg = 1556 # Jy
Ycyg = np.max(spec_fin)/np.min(spec_fin)
SEFD = Scyg/(Ycyg-1)

print(SEFD) # Jy
```

```python
# Gaussian function to be used with scipy curve_fit
def func(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
```

```python
transits = spec_fin.reshape(6, -1)
trans_times = times_cyga.reshape(6,-1)

FWHM = []

for i in range(6):
    t_ = trans_times[i]
    transit_ = transits[i]
    
    # only look at source (not bkg) + bring min to 0 to apply Gaussian fit and calculate FWHM
    tmin = np.min(transit_)
    idx = np.where(transit_ != np.min(transit_))
    transit = transit_[idx] - tmin
    t = t_[idx]
    
    # use scipy curvefit to solve for Gaussian profile parameters
    mean_guess = np.mean(t)
    sig_guess = np.std(t)
    a_guess = np.max(transit)
    popt, pcov = optimize.curve_fit(func, t, transit, p0=[a_guess,mean_guess,sig_guess])
    
    #make plot of transits
    plt.plot(t_, transit_)
    #plt.plot(t, tmin+func(t, popt[0], popt[1], popt[2]), 'k-')
    
    # calculate FWHM = 2*sqrt(2*ln(2))*sigma
    # slew rate = 0.57 deg/s
    sig = popt[2]
    FWHM.append(sig*0.57*2*np.sqrt(2*np.log(2)))
    
print(FWHM) # in degrees
plt.xlabel('Time (s)');
plt.ylabel('Power');
plt.title('Background Corrected Cyg A Transits');
```
